```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot(range(10), np.array(range(10))**2, '*y')
plt.show()
plt.plot(range(10), np.array(range(10))**3, '^g')
plt.show()
```

```python
plt.plot(range(10), np.array(range(10))**2, '*y', label = 'First')
plt.plot(range(10), np.array(range(10))**3, '^g', label = 'Second')
plt.legend()
plt.show()
```

```python
nums = np.random.randint(11,14,100)
s = pd.Series(nums)
print(s.head())
r = s.value_counts()
plt.bar(x  = r.index, height = r.values)
for idx in r.index:
    plt.text(idx, r[idx] +2, "[{}]".format(idx))
plt.show()

r.plot(kind='bar', legend  = True)
```

```python
nums = np.random.randint(11,14,100)
s = pd.Series(nums)
print(s.head())
r = s.value_counts()
# matplotlib
plt.pie(r.values, labels = r.index, explode = [0.0,0.0,0.4], shadow = True)
plt.legend()
plt.show()

r.plot(kind='pie')
```

```python
nums = np.random.randint(11,14,100)
s = pd.Series(nums)
print(s.head())
r = s.value_counts()
plt.barh(y= r.index, width = r.values)
plt.show()

r.plot(kind='barh', legend  = True)
```
