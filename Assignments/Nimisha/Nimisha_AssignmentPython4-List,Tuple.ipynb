#### 5. Print Elements at Odd indexes from a list (Do not use loop) 
    l = [10,11,20, 21,30, 31, 40, 41] 

```python
l = [10, 11, 20, 21,31, 40,41] 
  
odd = [num for num in l if num % 2 == 1] # using list comprehension
  
print(odd) 

#l[1::2]
```
