```python
#5.WAP to print Fibonacci sequence till n.(Fibonacci series is a series in which each number is the sum of the preceding two numbers)
n = int(input("Enter the value of n: "))
a = 0
b = 1
sum = 0
count = 1
while(count <= n):
    print(sum, end = " ")
    count += 1
    a = b
    b = sum
    sum = a + b
```

```python
#8.WAP to find sum of following series given n as input from user.1 + 2! + 3! + 4! + 5! + â€¦.n! Where n! denotes the factorial of number n.

N = int(input("Enter the Value of N:"))
S=0

for i in range(1, N+1):
    fact=1
    
    for j in range(1, i+1):
        fact=fact*j
    
    S=S+fact
    
    print(str(i)+"!",end="")
    if i!=N:
        print("+",end="")
print("=" + str(S))
```

```python
n=int(input("Enter the number of rows: "))
 
for i in range(1,n+1):
    
    for j in range(i,n):
        print(end=" ")
 
    for j in range(1,i+1):
        print("*",end="")
    print()
```
