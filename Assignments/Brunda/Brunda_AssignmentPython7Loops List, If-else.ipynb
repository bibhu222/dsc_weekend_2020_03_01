```python
#3.Write a function that takes a list of numbers from user as argument and returns the sum of only odd numbers (Use only for loop. No need to use if statement)

nl = []
Even_Sum = 0
Odd_Sum = 0

n = int(input("enter the number of elements: "))
for i in range(1, n + 1):
    value = int(input("enter the values"))
    nl.append(value)

for j in range(n):
    if(nl[j] % 2 == 0):
        Even_Sum = Even_Sum + nl[j]
    else:
        Odd_Sum = Odd_Sum + nl[j]

print("The Sum of Odd Numbers in this List =  ", Odd_Sum) 
```

```python
#4.WAP to input a list of numbers and store in a tuple. Now input another number and print the index of this number in the tuple. [Use index function] 

values = input("Input comma seprated numbers : ")
lis = values.split(",")
tu = tuple(lis)
print('Tuple : ',tu)
n=input("enter the new number:")
tu=tu+(n,)
print('new tuple:',tu)
m=tu.index(n)#index function
print("the index of the element {0} is {1}".format(n,m))
```

```python
#5.Create a list of 10 random numbers and then print the list, sum of all numbers and average. Use a for loop. Also print elements at only even indexes in the list.
#method1
import random

my_randoms=[]
for i in range (10):
    my_randoms.append(random.randrange(1,101,1))
print (my_randoms)

total =0
for ele in range(0, len(my_randoms)): 
    total = total + my_randoms[ele] 
    average = total/len(my_randoms)
print("the sum of the list is:",total)
print("the average of the list is:",average)
print("the  elements at only even indexes in the list rand",end=" ")
print(my_randoms[0:8:2])
```

```python
#8.WAP to print total number of spaces in a string (Use for loop to iterate over the string. Don’t use count function).

string = "I AM BRUNDA :)"
counter = 0
for i in string:
    if i == ' ' or i == '\n':
        counter += 1
print(string)
print(counter)
```
