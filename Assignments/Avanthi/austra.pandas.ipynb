```python
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
df.drop(df.loc[df['Type']=='Emigrants'].index, inplace = True)
```

```python
#Find columns for which more than 50% data is missing and drop them (Column Deletion)
df.isna().any()
```

```python
df.replace(0, np.nan, inplace = True)
df.head()
```

```python
df.replace('..', np.nan, inplace = True)
df.head()
```

```python
df.dropna(axis = 1, thresh = int(224 *(0.50)) , inplace = True)
df.columns
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
#Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) 
#Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"
df.drop(columns = ['Type', 'Coverage', 'AREA', 'REG', 'DEV'], inplace = True)
```

```python
df.columns
```

```python
# Rename columns :OdName -> Country; RegName -> Region
df.rename(columns = {'OdName': 'Country','RegName' : 'Region'}, inplace = True)
```

```python
# Convert Column Names with Int type to Strings to avoid index errors with string indexing 
#"1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
df.rename(columns = {1980 : '1980', 1981 :'1981', 1982 : '1982', 1983 :'1983', 1984:'1984', 1985: '1985', 1986:'1986', 
                     1987: '1987',1988:'1988', 1989: '1989', 1990:'1990', 1991:'1991', 1992:'1992', 1993:'1993', 
                     1994:'1994', 1995:'1995', 1996:'1996', 1997:'1997', 1998 :'1998', 1999:'1999',2000:'2000',
                     2001:'2001', 2002:'2002', 2003:'2003', 2004:'2004', 2005:'2005', 2006:'2006', 2007:'2007', 2008:'2008'},
                    inplace = True)
```

```python
df.drop(columns=['Total_in_df'], inplace = True)
```

```python
#Find Total number of immigrants for all years, and create a new column Total in data Frame
df['Total_in_df'] =df.sum(axis =1, skipna = True, numeric_only=True)
```
